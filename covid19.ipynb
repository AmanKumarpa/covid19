```python
covid_df2["Date"] = pd.to_datetime(covid_df["Date"], format = "%Y-%m-%d")
covid_df2.head()
```

```python
for col in ["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"]:
    if col in covid_df2.columns:
        covid_df2.drop(col, axis=1, inplace=True)
covid_df2
```

covid_df2.drop(["Sno", "Time" , "ConfirmedIndianNational","ConfirmedForeignNational"],
              axis = 1 , inplace = True)
covid_df2


```python
#Active Cases
covid_df2["Active_Cases"] = covid_df2["Confirmed"]-(covid_df2["Cured"]+covid_df2["Deaths"])
covid_df2
```

```python
statewise = pd.pivot_table (covid_df2, values = ["Confirmed", "Deaths", "Cured"],
                           index = "State/UnionTerritory", aggfunc = max)
statewise
```

```python
statewise["Recovery Rate"]= statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"]= statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise =  statewise.sort_values(by = "State/UnionTerritory"  )
```

```python
style = statewise.style.background_gradient(cmap = "viridis")
style
```

```python
#remove rows where state = india
vaccine1 = vaccine_df[vaccine_df.State !="India"]
vaccine1 
```

```python
vaccine2 = vaccine1.rename(columns = {"Total Individuals Vaccinated" : "Total"}, inplace = False)

vaccine2
```

```python
#most vaccinated states
max_vac = vaccine2.groupby( "State")["Total"].sum().to_frame()

max_vac = max_vac.sort_values ( "Total", ascending = False )[:5]
max_vac
```

```python
plt.figure(figsize = (10,5))
plt.title("Top 5 Vaccinated States in India")
plt.xlabel("States")
plt.ylabel("Total")
sns.barplot(x = "State", y = "Total", data = max_vac, palette ="CMRmap" , linewidth  = 2,
            edgecolor ="b" , )
plt.xticks (rotation = 60)
plt.show()
          
```

```python
#least vaccinated states
min_vax = vaccine2.groupby("State")["Total"].sum().to_frame()
min_vax.sort_values("State").min()
min_vax.head(5)
```
